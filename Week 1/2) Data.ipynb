## Data Understanding

In order to develop a model, data on the aforementioned causes behind a car collision is needed. In 2010, the city of Seattle launched the "Open Data Program" which made the data generated by the City of Seattle publicly available with the purpose of increasing the quality of life of its residents[[4](http://www.seattle.gov/tech/initiatives/open-data/about-the-open-data-program)]. 
In this project, the data made available by the City of Seattle will be used in order to create a model that can predict the probability of a driver getting into a collision (given road and weather conditions) as well as the possible severity of the predicted accident.

The city of Seattle is the largest in King County, and with a total population of 753,675 [[5](https://www.census.gov/quickfacts/fact/table/seattlecitywashington,US/PST045219)] it also holds the title for largest city in the State of Washington. According to the Washington State Department of Transit, a total number of 40,962 collisions occured in 2015 in King County, WA [[6](https://www.wsdot.wa.gov/mapsdata/crash/pdf/2015_Annual_Collision_Summary.pdf)]. The amount of publicly available information of Seattle makes this city a good subject of study for developing this project's target model.

There are 221,525 collisions registered by the Seattle Department of Transit (from now on: SDOT) from January 1st 2004 up to September 15th, 2020 [[7](https://data.seattle.gov/Land-Base/Collisions/9kas-rb8d)]. Among the information stored by the SDOT we can find: the location of the accident (X and Y coordinates), the severity of the collision (Property damage, injury, serious injury, and fatalitiy), weather (clear, fog, overcast, etc.), and road conditions(dry, ice, oil, etc.) at the time of the accident, and so much more.

Using the [Metadata](https://www.seattle.gov/Documents/Departments/SDOT/GIS/Collisions_OD.pdf "SDOT COLLISIONS") file that describes the data collisions attributes above, redundant variables can be delated (i.e 'Severity Code' is just the numerical representation of 'Severity Description'), and a model to predict the seveirty of a collision based on different attributes can be made. After developing a model, testing to determine accuracy will be done. For the first approach to his model, data on coordinates, SDOT collision code, weather, road, and light conditions will be used.
