## Business understanding

The number of car crashes in the US have fallen since 2005; nonetheless, motor vehicle crashes are still the leading cause of death for Americans under 30 [[1](https://www.cdc.gov/winnablebattles/report/motor.html)]. Furthermore, the US Department of Transportation estimated in 2010 that car collisions took 242 billion USD off the American economy [[2](https://crashstats.nhtsa.dot.gov/Api/Public/ViewPublication/812013)]  due to lost of productivity, medical costs, legal costs, emergency services, insurance, property damage, etc. 

According to Rolison *et al*.[[3](https://doi.org/10.1016/j.aap.2018.02.025)], the main factors behind a car collision are: failure to look properly, loss of control, failure to judge another person's path or speed, traveling too fast for conditions, slippery roads, and carelessness. Knowing the causes behind a car collision could help to make sure they do not happen in the first place. Preventing vehicle collisions could not only help to avoid the loss of many young American lives (and the emotional burden that brings to their loved ones), but also save billions of dollars to the US economy. In other words, preventing vechicle collisions will improve the quality of life of millions of people.

The objective of this project is to use machine learning to develop a model that can warn the driver for the posibility of them getting into a car accident, and its severity, given the weather and road conditions. This could help the driver pay more attention when driving in areas where the probability of an accident is higher, or even change the route to a safer one. 

## Data Understanding

In order to develop a model, data on the aforementioned causes behind a car collision is needed. In 2010, the city of Seattle launched the "Open Data Program" which made the data generated by the City of Seattle publicly available with the purpose of increasing the quality of life of its residents[[4](http://www.seattle.gov/tech/initiatives/open-data/about-the-open-data-program)]. 
In this project, the data made available by the City of Seattle will be used in order to create a model that can predict the probability of a driver getting into a collision (given road and weather conditions) as well as the possible severity of the predicted accident.

The city of Seattle is the largest in King County, and with a total population of 753,675 [[5](https://www.census.gov/quickfacts/fact/table/seattlecitywashington,US/PST045219)] it also holds the title for largest city in the State of Washington. According to the Washington State Department of Transit, a total number of 40,962 collisions occured in 2015 in King County, WA [[6](https://www.wsdot.wa.gov/mapsdata/crash/pdf/2015_Annual_Collision_Summary.pdf)]. The amount of publicly available information of Seattle makes this city a good subject of study for developing this project's target model.
